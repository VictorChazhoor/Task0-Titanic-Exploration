```python
#importing modules pandas and numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
```

```python
#loading dataset using pandas stored in variable ds
ds = pd.read_csv("train.csv")
ds
```

```python
#important information about dataset
ds.info()
```

```python
#Removing all rows containing null value
ds1 = ds.dropna()
ds1
```

```python
#dropping column PassengerId and Name
ds2 = ds.drop(columns=['PassengerId', 'Name'])
ds2
```

```python
#dividing dataset in ratio 80:20
n = ds.shape[0]
n1 = int(0.8*n)+1
ds3 = ds.iloc[0:n1]
ds3
```

```python
ds4 = ds.iloc[n1:n]
ds4
```

```python
#Histogram for column fare
pl.hist(ds1['Fare'])
pl.title('Fare')
pl.show()
```

```python
#Histogram for column age
pl.hist(ds1['Age'])
pl.title('Age')
pl.show()
```

```python
#Bar Chart for Column Survived
c1 = 0
c0 = 0
for i in ds1['Survived']:
    if i:
        c1+=1
    else:
        c0+=1
x = np.array(['Deceased', 'Survived'])
y = np.array([c0, c1])
pl.bar(x, y)
pl.title('Survived Bar Chart')
pl.ylabel('No of people')
pl.show()
```

```python
#Bar Chart for Column Sex
c1 = 0
c0 = 0
for i in ds1['Sex']:
    if i=='female':
        c1+=1
    else:
        c0+=1
x = np.array(['Male', 'Female'])
y = np.array([c0, c1])
pl.bar(x, y, color=['blue', 'pink'])
pl.title('Sex Bar Chart')
pl.ylabel('No of people')
pl.show()
```
